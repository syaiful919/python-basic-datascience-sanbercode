### Jawab
####  Apa itu machine learning?
machine learning merupakan salah satu cabang dari Artificial Intelligence, dimana mesin (komputer) dapat belajar dari data-data sehingga kinerjanya dapat meningkat

#### Apa itu data feature dan data target
data feature adalah predictor sedangkan data target adalah output yang akan diprediksi

#### Apa Perbedaan Supervised Learning, dan Unsupervised Learning?
Pada supervised learning, dataset yang digunakan memiliki label, sedangkan pada unsupervised learning, dataset yang digunakan tidak memiliki label.

#### Apa Jenis2 yang ada di dalam Supervised Learning? Jelaskan Perbedaannya!
terdapat 2 jenis, yaitu klasifikasi & regresi. pada klasifikasi, outputnya adalah class label, sedangkan pada regresi, outpputnya adalah rentang value yang kontinu

<h1>Quiz 2 : Pemahaman Machine Learning</h1>

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian applikasikan untuk membuat model bagi data (variable x, y) di bawah ini. kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeRegressor

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

x = np.reshape(x, (-1,1))

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)
x_new = np.reshape(x_new, (-1,1))
x_new

regr = DecisionTreeRegressor(max_depth=2)
regr.fit(x, y)

y_new = regr.predict(x_new)
y_train = regr.predict(x)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y, label="Data Asli")
ax.plot(x_new, y_new, color="green", label="Prediksi terhadap Data Baru", linewidth=2)
ax.scatter(x, y_train, marker='x', color='yellow', label="Prediksi terhadap Data Training")

ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
ax.legend(bbox_to_anchor=(1.2, 1.2))
plt.show()
```

```python
regr.get_params()
```

parameters:
- ccp_alpha : 0.0
- criterion : mse
- max depth : 2

```python
from sklearn import tree

tree.plot_tree(regr, filled = True)
```
