2. Beberapa cara imputing missing value
   - drop missing value, biasanya dilakukan jika sebuah feature memiliki terlalu banyak data yang kosong
   - Diisi dengan nilai mean/median, biasanya dilakukan jika datanya bertipe numerik
   - Diisi dengan modus, biasanya dilakukan untuk data bertipe kategori
   - Diisi dengan data sebelum atau sesudahnya, biasanya dilakukan untuk data time series
   - Diisi dengan algoritma KNN
3. Jika algoritma model yang digunakan adalah distance based
4. Mengumpulkan data -> pre processing data (EDA & feature engineering) -> modeling (membuat model & optimasi, training & evaluasi) -> deploy -> monitoring

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import roc_auc_score
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

```python
# column cabin memiliki terlalu banyak missing value, lebih baik di drop
df = df.drop('Cabin', axis= 1)
```

```python
# melihat modus dari column embarked
df['Embarked'].value_counts()
```

```python
# mengsisi missing value column embarked dengan nilai modusnya
df['Embarked'] = df['Embarked'].fillna('S')

# ubah nilainya menjadi numeric
embarked_dict = {'S':0, 'C':1, 'Q': 2}
df['Embarked'] = df['Embarked'].map(embarked_dict)
```

```python
# mengisi missing value column age dengan nilai rata-ratanya
df['Age'] = df['Age'].fillna(np.round_(df['Age'].mean()))
```

```python
# mengubah nilai column sex menjadi numeric
sex_dict = {'male':0, 'female':1}
df['Sex'] = df['Sex'].map(sex_dict)
```

```python
# mendrop column yang tidak diperlukan
df = df.drop(['PassengerId', 'Name', 'Ticket'], axis= 1)
```

```python
# standarisasi column age dan fare

std_scaller = StandardScaler()
df[['Age']] = std_scaller.fit_transform(df[['Age']])
df[['Fare']] = std_scaller.fit_transform(df[['Fare']])
```

```python
# modeling

X = df.drop('Survived', axis= 1)
y = df['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.25, random_state= 21, stratify= y)

model_knn = KNeighborsClassifier()
param_grid = {
    'n_neighbors': np.arange(5, 30),
    'weights': ['uniform', 'distance']
}
rscv = RandomizedSearchCV(model_knn, param_grid, cv=10, scoring= 'roc_auc', n_iter=50)
rscv.fit(X_train, y_train)
```

```python
rscv.best_params_
```

```python
rscv.best_score_
```

```python
y_pred_survived = rscv.predict_proba(X_test)
```

```python
rscv.classes_
```

```python
y_survived_pred = y_pred_survived[:, 1]
```

```python
roc_auc_score(y_test, y_survived_pred)
```
